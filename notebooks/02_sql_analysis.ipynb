```python
#Import

import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect("../data/sales.db")
```

```python
# Query to list all tables available in the SQLite database
query = """
SELECT name 
FROM sqlite_master 
WHERE type='table';
"""

# Displaying the result as a DataFrame
pd.read_sql(query, conn)
```

```python
# SQL query to calculate total revenue across all sales
query = """
SELECT 
    SUM(revenue) AS total_revenue
FROM sales;
"""

# Executing the query and displaying the result
pd.read_sql(query, conn)
```

```python
# SQL query to calculate total revenue grouped by product
query = """
SELECT 
    product,
    SUM(revenue) AS total_revenue
FROM sales
GROUP BY product
ORDER BY total_revenue DESC;
"""

# Displaying revenue performance by product
pd.read_sql(query, conn)
```

```python
# SQL query to calculate total revenue grouped by region
query = """
SELECT 
    region,
    SUM(revenue) AS total_revenue
FROM sales
GROUP BY region
ORDER BY total_revenue DESC;
"""

# Displaying revenue performance by region
pd.read_sql(query, conn)
```

```python
# Closing the database connection to free resources
conn.close()
```
